# **Project Summary**

## **1. Business Case**

**INX Future Inc. (INX)** is a leading data analytics and automation solutions provider with over 15 years of global presence. In recent years, the organization has observed a decline in employee performance indices, which has become a significant concern for top management. This decline has resulted in increased service delivery escalations and an 8% drop in client satisfaction levels.

To address this issue, the CEO has initiated a data-driven analysis of employee performance to identify the underlying causes. The organization also aims to develop a mechanism to identify non-performing employees in a way that minimizes negative impact on overall employee morale.

### **Key Objectives**

* Analyze department-wise performance trends
* Identify the top factors influencing employee performance
* Build a predictive model to forecast employee performance
* Provide actionable recommendations to improve performance

#### **Univariate Analysis**

* The average employee age is **36**, with most employees between **25–45 years**
* A majority of employees live close to the workplace
* Most employees rated their job environment as **3.0**
* Employees typically have experience in **0–1 prior companies**

#### **Bivariate Analysis**

* The **Development** department shows the highest performance, followed by **Research & Development**
* Roles such as **Business Analyst, Technical Lead, Data Scientist, and Developer** demonstrate strong performance
* Employees who **work overtime** tend to have higher performance ratings
* Performance differences across gender appear influenced by **data imbalance** rather than clear causal factors
* Employees from **Life Sciences and Medical backgrounds** show relatively higher performance
* Employees who **travel rarely** perform better than frequent travelers
* Performance declines when employees remain with the **same manager for extended periods**
* Lack of promotion for **2–3 years** negatively impacts performance
* Strong **work-life balance, relationship satisfaction, and environment satisfaction** correlate with higher performance
* Employees with **>10 years of experience** show a slight decline in performance
* Salary hikes above **19%** significantly improve performance
* Approximately **11%–16.3%** of employees with low job satisfaction and involvement exhibit low performance

## **4. Data Preprocessing**

* **Missing Values:** No missing values found
* **Encoding:** Categorical variables transformed using **Label Encoding**
* **Outlier Handling:**

  * **IQR method** is used handle the outliers
* **Feature Transformation:**

  * Applied **log transformation** to reduce the **high Skewness**
* **Feature Scaling:**

  * Applied **Standard Scaling** (mean = 0, standard deviation = 1)



## **5. Feature Selection**

* Correlation analysis was used to identify important predictors
* Selected features include:

  * Department
  * Job Role
  * Environment Satisfaction
  * Last Salary Hike Percent
  * Work-Life Balance
  * Experience Years (Company & Role)
  * Years Since Last Promotion
  * Years with Current Manager

### **Top 3 Key Factors Affecting Performance**

1. **Environment Satisfaction**
2. **Last Salary Hike Percentage**
3. **Years Since Last Promotion**

* Removed irrelevant features such as **Employee Number**
* Two multicollinearity issues were identified and resolved by combining the correlated features into a new feature and removing the original variables.
* No duplicate records found



## **6. Model Development & Evaluation**

### **Steps Performed**

* Defined dependent and independent variables
* Addressed class imbalance using **SMOTE (Synthetic Minority Oversampling Technique)**
* Split data into **75% training** and **25% testing**
* Implemented a Pipeline to streamline preprocessing and modeling
* Pipeline Benefits:

1. Prevents data leakage by applying transformations only on training data
2. Ensures consistent preprocessing across training and testing
3. Improves reproducibility and scalability of the model
4. Simplifies workflow by combining all steps into a single process

### **Algorithms Evaluated**

* Logistic Regression
* Support Vector Machine (SVM)
* Decision Tree
* Random Forest
* Artificial Neural Network (MLP Classifier)

### **Model Performance Summary**

* **Logistic Regression:** Lower performance, indicating underfitting (Train: ~77%, Test: ~76%)
* **SVM:** High training accuracy with noticeable overfitting (Train: ~97%, Test: ~85%)
* **Decision Tree:** Good performance but prone to overfitting (Train: ~97%, Test: ~88%)
* **Random Forest:** **Best and most stable model with strong generalization** **(Train: ~96%, Test: ~93%)**
* **ANN (MLP):** High capacity but overfitting observed (Train: ~98%, Test: ~84%)

Despite some models achieving higher training accuracy, the **tuned Random Forest model** provides the best overall performance with superior test accuracy and a well-balanced bias–variance trade-off, making it the most reliable and robust choice for this dataset.


## **7. Recommendations**

Based on the analysis and model insights, the following actions are recommended:

* Enhance **employee environment satisfaction**, as it has the strongest impact on performance
* Introduce **manager rotation every 2–3 years** to ensure sustained employee engagement and productivity
* Offer **competitive salary hikes**, particularly above 19%, to improve motivation and retention
* Maintain a **balanced promotion cycle** (preferably 1–2 years) to ensure realistic and sustainable growth
* Strengthen **work-life balance initiatives** to reduce burnout and improve overall performance
* Proactively identify and support employees with **low job and relationship satisfaction**
* Focus on optimizing performance in **underperforming departments**, while sustaining excellence in high-performing ones like Development and Sales
* Leverage the **predictive model** in hiring and performance management to identify high-potential and at-risk employees early.
