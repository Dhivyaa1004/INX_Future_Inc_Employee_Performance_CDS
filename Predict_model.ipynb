## **LOAD  & PREDICT THE MODEL**

```python
import joblib

pipeline = joblib.load('Random_Forest_Tuned_Model.pkl')
```

```python
import pandas as pd

new_data_1 = pd.DataFrame({
    'Age': [35],
    'Gender': ['Male'],
    'EducationBackground': ['Other'],
    'MaritalStatus': ['Single'],
    'EmpDepartment': ['Sales'],
    'EmpJobRole': ['Sales Representative'],
    'BusinessTravelFrequency': ['Non-Travel'],
    'DistanceFromHome': [30],
    'EmpEducationLevel': [1],
    'EmpEnvironmentSatisfaction': [1],
    'EmpHourlyRate': [40],
    'EmpJobInvolvement': [2],
    'EmpJobLevel': [4],
    'EmpJobSatisfaction': [1],
    'NumCompaniesWorked': [2],
    'OverTime': ['No'],
    'EmpLastSalaryHikePercent': [15],
    'EmpRelationshipSatisfaction': [2],
    'TotalWorkExperienceInYears': [8.0],
    'TrainingTimesLastYear': [2],
    'EmpWorkLifeBalance': [1],
    'ExperienceYearsAtThisCompany': [4.0],
    'ExperienceYearsInCurrentRole': [2],
    'YearsSinceLastPromotion': [1.0],
    'Attrition': ['Yes']
})   

#  Create ExperienceRatio
new_data_1['ExperienceRatio'] = (
    new_data_1['ExperienceYearsInCurrentRole'] /
    new_data_1['TotalWorkExperienceInYears']
)

prediction = pipeline.predict(new_data_1)

mapping = {1: "Low", 2: "Good", 3: "Excellent", 4: "Outstanding"}

print("Predicted Rating:", prediction[0])
print("Category:", mapping[prediction[0]])
```
