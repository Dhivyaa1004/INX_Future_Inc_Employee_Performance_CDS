### HANDLING THE OUTLIERS
QR-based outlier treatment was chosen as it effectively handles extreme values while preserving the underlying data distribution, making it more suitable for model performance compared to winsorization.

```python
# ==== Defining the function for outlier detection and percentage calculation using IQR ====

def detect_outliers(data):
    datas = np.array(data)

    # ==== Quartiles ====
    q1 = np.quantile(datas, 0.25)
    q2 = np.quantile(datas, 0.50)
    q3 = np.quantile(datas, 0.75)

    # ==== IQR & boundsa ====
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR

    # ==== Outlier detection ====
    outliers = datas[(datas < lower_bound) | (datas > upper_bound)]
    outlier_count = len(outliers)
    outlier_percent = round(outlier_count * 100 / len(datas), 2)

    # ==== Display results ====
    print(f"Q1 = {q1}, Q2 (Median) = {q2:.2f}, Q3 = {q3}")
    print(f"IQR = {IQR:.2f}")
    print(f"Lower Bound = {lower_bound:.2f}, Upper Bound = {upper_bound:.2f}")
    print(f"Outliers Detected: {outlier_count}")
    print(f"Outlier Percentage: {outlier_percent}%\n")

    # ==== Calculating IQR, Lower/Upper Bounds, and Outlier Counts for Continuous Numerical Features ====

# Get only numerical columns to apply outlier detection
for feature in data.select_dtypes(include=np.number).columns:
  print(feature,":")
  detect_outliers(data[feature])
  print("-"*50)
```

```python
col_to_clip = ['NumCompaniesWorked','TotalWorkExperienceInYears','TrainingTimesLastYear','ExperienceYearsAtThisCompany','ExperienceYearsInCurrentRole', 'YearsSinceLastPromotion','YearsWithCurrManager']

# Keep original data
data_before = data.copy()
data_after = data.copy()

def outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    data[col] = data[col].clip(lower, upper)
    return data

# Apply clipping only to the specified columns
for feature_name in col_to_clip:
    data = outliers(data, feature_name)
```

#### Temporary Label Encoding for Correlation Analysis

Temporary label encoding was applied to categorical variables solely for the purpose of correlation analysis. This enabled the computation of relationships between features and the target variable (PerformanceRating). 

Based on the correlation results, the top three features most strongly associated with employee performance were identified. This encoding was used only for analysis and was not included in the final model pipeline to avoid introducing bias.

```python
# ===================== for the temporary label encoding  for checking the correlation=======================
from sklearn.preprocessing import LabelEncoder
temp_data = data.copy()
cat_cols=['Gender', 'EducationBackground', 'MaritalStatus',
       'EmpDepartment', 'EmpJobRole', 'BusinessTravelFrequency', 'OverTime',
       'Attrition']
encoder = LabelEncoder()
# Loop through the columns in cat_cols1 and apply LabelEncoder
for col in cat_cols:
    temp_data[col] = encoder.fit_transform(temp_data[col])
temp_data.head(2)
```

```python
#================Finding out the correlation coeffecient to find out which predictors are significant. =====================
core=temp_data.corr()
core[core>0.8]
```

```python
# =========== Identifying Highly Correlated Features ===============
import numpy as np

corr_matrix = temp_data.corr()

# Upper triangle (avoid duplicate pairs)
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find columns to drop (correlation > 0.8)
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]

print(to_drop)
```

Highly correlated features such as ExperienceYearsInCurrentRole and YearsWithCurrManager were removed to reduce multicollinearity and redundancy, ensuring better model performance and interpretability.

```python
# ========== Create a new feature for control the correlation =================
data['ExperienceRatio'] = data['ExperienceYearsInCurrentRole'] / data['TotalWorkExperienceInYears']
```

Feature combination means creating a new variable by combining related features to capture meaningful relationships and reduce redundancy.

* **ExperienceRatio tells: How much of total experience is spent in current role.**

```python
# ===========To remove highly correlated (redundant) features ======================
data = data.drop([
    'ExperienceYearsInCurrentRole',
    'YearsWithCurrManager'
], axis=1)
```

```python
# ============ Check and remove the null values using median ===============
import numpy as np

data['ExperienceRatio'].replace([np.inf, -np.inf], np.nan, inplace=True)
data['ExperienceRatio'].fillna(data['ExperienceRatio'].median(), inplace=True)
```

Invalid values generated during feature engineering were first handled by converting them to NaN, followed by median imputation to ensure clean and reliable feature representation.
